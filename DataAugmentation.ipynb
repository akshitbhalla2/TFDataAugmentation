```python
import os
import numpy as np
import tensorflow as tf
from PIL import Image
from matplotlib import pyplot as plt
```

# Rotation

```python
image_path = 'images/train/cat/cat.jpg'

plt.imshow(plt.imread(image_path));
```

```python
generator = tf.keras.preprocessing.image.ImageDataGenerator(
    rotation_range = 40 # Within 40 clock or anticlock
)

x, y = next(generator.flow_from_directory('images', batch_size=1))
plt.imshow(x[0].astype('uint8'));
```

# Width and Height Shifts

```python
generator = tf.keras.preprocessing.image.ImageDataGenerator(
    width_shift_range = [-100, -50, 0, 50, 100], # Only numbers in list
    height_shift_range = (-50, 50) # Any number between -50 and 50
)

x, y = next(generator.flow_from_directory('images', batch_size=1))
plt.imshow(x[0].astype('uint8'));
```

# Brightness

```python
generator = tf.keras.preprocessing.image.ImageDataGenerator(
    brightness_range = (0.5, 2)
)

x, y = next(generator.flow_from_directory('images', batch_size=1))
plt.imshow(x[0].astype('uint8'));
```

# Shear Transformation

```python
generator = tf.keras.preprocessing.image.ImageDataGenerator(
    shear_range = 40 # Within 40 clock or anticlock
)

x, y = next(generator.flow_from_directory('images', batch_size=1))
plt.imshow(x[0].astype('uint8'));
```

# Zoom

```python
generator = tf.keras.preprocessing.image.ImageDataGenerator(
    zoom_range = -0.5 # Zoom in or out
)

x, y = next(generator.flow_from_directory('images', batch_size=1))
plt.imshow(x[0].astype('uint8'));
```

# Channel Shift

```python
generator = tf.keras.preprocessing.image.ImageDataGenerator(
    channel_shift_range = 100
)

x, y = next(generator.flow_from_directory('images', batch_size=1))
plt.imshow(x[0].astype('uint8'));
```

# Flips

```python
generator = tf.keras.preprocessing.image.ImageDataGenerator(
    horizontal_flip = True,
    vertical_flip = True,
)

x, y = next(generator.flow_from_directory('images', batch_size=1))
plt.imshow(x[0].astype('uint8'));
```
